```python
# ============================================================
# STEP 6: DATA PREPROCESSING
# ============================================================

from sklearn.preprocessing import LabelEncoder    # Import LabelEncoder for converting categorical variables to numeric

def preprocess_data(df):
    """Preprocess the Student Performance dataset"""

    df_processed = df.copy()    # Create a copy to avoid changing the original dataset

    # -----------------------------
    # Handle missing values
    # -----------------------------
    # Fill missing values in numerical columns with median
    numeric_cols = df_processed.select_dtypes(include=['int64', 'float64']).columns
    for col in numeric_cols:
        df_processed[col].fillna(df_processed[col].median(), inplace=True)

    # Fill missing values in categorical columns with mode
    categorical_cols = df_processed.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        df_processed[col].fillna(df_processed[col].mode()[0], inplace=True)

    # -----------------------------
    # Convert categorical variables to numeric
    # -----------------------------
    label_encoder = LabelEncoder()
    for col in categorical_cols:
        df_processed[col] = label_encoder.fit_transform(df_processed[col])

    # -----------------------------
    # Drop unnecessary columns if any (optional)
    # -----------------------------
    # For this dataset, we may not have unnecessary columns like Name or ID,
    # but you can drop them if present
    columns_to_drop = ['student_name', 'id']   # Example columns, adjust as per your dataset
    df_processed = df_processed.drop([col for col in columns_to_drop if col in df_processed.columns], axis=1)

    return df_processed

# -----------------------------
# Preprocess training and test data
# -----------------------------
train_processed = preprocess_data(train_df)    # Preprocess the training dataset
test_processed = preprocess_data(test_df)      # Preprocess the test dataset

# Print shapes of processed datasets
print("Processed training set shape:", train_processed.shape)
print("Processed test set shape:", test_processed.shape)

# Display first few rows of processed training data
train_processed.head()
```

```python
# ============================================================
# STEP 7: PREPARE DATA FOR MODELING
# ============================================================

from sklearn.model_selection import train_test_split    # Import function to split dataset
from sklearn.preprocessing import StandardScaler        # Import scaler for feature normalization

# -----------------------------
# Split features and target
# -----------------------------
# Replace 'target' with the actual target column in your dataset, e.g., 'math score' or a performance label
target_column = 'math score'    # Example: you can choose 'math score', 'reading score', or a custom target
X = train_processed.drop(target_column, axis=1)   # Features = all columns except target
y = train_processed[target_column]                # Target variable

# -----------------------------
# Split data into training and validation sets
# -----------------------------
# 80% training, 20% validation; stratify optional if target is categorical
X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=0.2,          # 20% of data for validation
    random_state=42         # For reproducibility
)

# -----------------------------
# Scale the features
# -----------------------------
scaler = StandardScaler()            # Create a StandardScaler object
X_train_scaled = scaler.fit_transform(X_train)   # Fit scaler on training data and transform
X_val_scaled = scaler.transform(X_val)           # Transform validation data using the same scaler

# -----------------------------
# Print shapes of training and validation sets
# -----------------------------
print(f"Training set features: {X_train.shape}")
print(f"Training set target: {y_train.shape}")
print(f"Validation set features: {X_val.shape}")
print(f"Validation set target: {y_val.shape}")
```

```python
# ============================================================
# STEP 8: CREATE AND TRAIN MULTIPLE REGRESSION MODELS
# ============================================================

from sklearn.linear_model import LinearRegression           # Linear Regression
from sklearn.tree import DecisionTreeRegressor              # Decision Tree Regressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor  # Ensemble regressors
from xgboost import XGBRegressor                             # XGBoost Regressor
from sklearn.metrics import mean_squared_error, r2_score    # Metrics for regression

# -----------------------------
# Initialize regression models
# -----------------------------
models = {
    'Linear Regression': LinearRegression(),                  # Linear regression
    'Decision Tree': DecisionTreeRegressor(random_state=42),  # Decision tree regressor
    'Random Forest': RandomForestRegressor(random_state=42),  # Random forest regressor
    'Gradient Boosting': GradientBoostingRegressor(random_state=42), # Gradient boosting regressor
    'XGBoost': XGBRegressor(random_state=42, eval_metric='rmse')      # XGBoost regressor
}

# -----------------------------
# Train models and store results
# -----------------------------
results = {}    # Dictionary to store model, metrics, and predictions

for name, model in models.items():

    # Fit the model on training data
    model.fit(X_train, y_train)          # Train on raw training features
    y_pred = model.predict(X_val)        # Predict on validation features

    # Evaluate the model
    mse = mean_squared_error(y_val, y_pred)    # Mean squared error
    r2 = r2_score(y_val, y_pred)               # R-squared score

    # Store results in dictionary
    results[name] = {
        'model': model,       # Trained model object
        'mse': mse,           # Mean squared error
        'r2': r2,             # R-squared score
        'predictions': y_pred # Predictions on validation set
    }

    # Print evaluation metrics
    print(f"{name}: MSE={mse:.4f}, R2={r2:.4f}")
```

```python
# ============================================================
# STEP 9: EVALUATE AND COMPARE MODELS
# ============================================================

import pandas as pd           # Import pandas to create comparison dataframe
import matplotlib.pyplot as plt  # Import matplotlib for plotting
import seaborn as sns         # Import seaborn for visualization

# -----------------------------
# Create a comparison dataframe
# -----------------------------
comparison_df = pd.DataFrame({
    'Model': list(results.keys()),                     # List of model names
    'MSE': [results[model]['mse'] for model in results.keys()],   # Mean Squared Error for each model
    'R2': [results[model]['r2'] for model in results.keys()]      # R-squared score for each model
}).sort_values('R2', ascending=False)                 # Sort by R2 descending (higher is better)

# Display the comparison table
print("Model Performance Comparison:")
print(comparison_df)

# -----------------------------
# Visualize model performance (R2)
# -----------------------------
plt.figure(figsize=(12, 6))                          # Set figure size
sns.barplot(data=comparison_df, x='R2', y='Model', palette='viridis')  # Horizontal bar plot of R2
plt.title('Model R2 Score Comparison')               # Set plot title
plt.xlim(0, 1)                                       # Limit x-axis from 0 to 1
plt.xlabel('R2 Score')                               # Label x-axis
plt.ylabel('Model')                                  # Label y-axis
plt.tight_layout()                                   # Adjust layout to prevent overlap
plt.show()

# -----------------------------
# Visualize model performance (MSE)
# -----------------------------
plt.figure(figsize=(12, 6))                          # Set figure size
sns.barplot(data=comparison_df, x='MSE', y='Model', palette='magma')  # Horizontal bar plot of MSE
plt.title('Model MSE Comparison')                    # Set plot title
plt.xlabel('Mean Squared Error')                     # Label x-axis
plt.ylabel('Model')                                  # Label y-axis
plt.tight_layout()                                   # Adjust layout
plt.show()
```

```python
# Identify the best model
# -----------------------------
# Choose best model based on highest R2 score from comparison_df
best_model_name = comparison_df.iloc[0]['Model']           # Get model name with highest R2
best_model = results[best_model_name]['model']             # Get trained model object
best_predictions = results[best_model_name]['predictions'] # Predictions on validation set

# Print basic information
print(f"Best Model: {best_model_name}")
print(f"Validation R2 Score: {results[best_model_name]['r2']:.4f}")
print(f"Validation MSE: {results[best_model_name]['mse']:.4f}")

# -----------------------------
# Detailed evaluation: Plot Actual vs Predicted
# -----------------------------
plt.figure(figsize=(8, 6))
plt.scatter(y_val, best_predictions, alpha=0.7)          # Scatter plot of actual vs predicted
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--') # Diagonal line for perfect prediction
plt.xlabel('Actual Values')                               # Label x-axis
plt.ylabel('Predicted Values')                            # Label y-axis
plt.title(f'Actual vs Predicted - {best_model_name}')    # Title of plot
plt.tight_layout()
plt.show()

# -----------------------------
# Residual Analysis: Plot residuals
# -----------------------------
residuals = y_val - best_predictions                     # Compute residuals (errors)
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, bins=30)               # Histogram of residuals
plt.xlabel('Residuals')                                  # Label x-axis
plt.ylabel('Frequency')                                  # Label y-axis
plt.title(f'Residual Distribution - {best_model_name}') # Title
plt.tight_layout()
plt.show()
```

```python

from sklearn.model_selection import cross_val_score    # Import cross-validation function
# -----------------------------
# Select top 3 models based on R2 score
# -----------------------------
top_models = comparison_df.head(3)['Model'].tolist()    # List of top 3 models
cv_results = {}                                        # Dictionary to store cross-validation scores

# -----------------------------
# Perform 5-fold cross-validation for each top model
# -----------------------------
for model_name in top_models:
    model = results[model_name]['model']               # Get the trained model object

    # Tree-based models and linear regression can use raw features
    # Use cross_val_score with R2 metric for regression
    cv_scores = cross_val_score(model, X, y, cv=5, scoring='r2')

    # Store cross-validation scores in dictionary
    cv_results[model_name] = cv_scores

    # Print cross-validation results
    print(f"{model_name} - Cross-validation R2 scores: {cv_scores}")
    print(f"{model_name} - Mean CV R2: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})\n")

# -----------------------------
# Visualize cross-validation results
# -----------------------------
plt.figure(figsize=(10, 6))                             # Set figure size
cv_df = pd.DataFrame(cv_results)                         # Convert dictionary to DataFrame for plotting
sns.boxplot(data=cv_df, palette='Set2')                 # Boxplot of R2 scores for each model
plt.title('Cross-Validation R2 Scores for Top Models')  # Set title
plt.ylabel('R2 Score')                                  # Label y-axis
plt.xticks(rotation=45)                                  # Rotate x-axis labels
plt.tight_layout()                                      # Adjust layout to prevent overlap
plt.show()
```
